```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv('./data.csv', delimiter=';')
df.info()
```

```python
df.head()
```

```python
df.columns = ['city', 'date', 'amount']
```

```python
df
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df
```

```python
df['month'] = df['date'].dt.month
df['year'] = df['date'].dt.year
df['day'] = df['date'].dt.day
```

```python
df
```

```python
df[df['amount'] > 0]
```

```python
df_grouped_by_date_amount = df.groupby(['date']).mean()
df_grouped_by_date_amount.drop(['month', 'year', 'day'], inplace=True, axis='columns')
df_grouped_by_date_amount
```
